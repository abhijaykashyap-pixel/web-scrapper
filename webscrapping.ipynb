```python
import pandas as pd
import requests
import time
from bs4 import BeautifulSoup
```

```python
headers = {"User-Agent": (
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) "
        "Chrome/120.0.0.0 Safari/537.36")}
webpage=requests.get('https://www.imdb.com/chart/top/',headers = headers).text
```

```python
soup=BeautifulSoup(webpage,'lxml')
movies = soup.find_all("h3", class_="ipc-title__text")
```

```python
name=[]
rating=[]
for i in movies:
    name.append(i.get_text(strip=True))
print(name)    
```

```python
r=soup.find_all("span",class_="ipc-rating-star--rating")    
for i in r:
    rating.append(i.text)
print(rating)    
```

```python
d=[]
dr=soup.find_all("span",class_="ipc-rating-star--voteCount")
for i in dr:
    d.append(i.text.strip())
print(d)    
```

```python
dictionary={'Movie Name':pd.Series(name), 'rating':pd.Series(rating), 'Vote Count':pd.Series(d) }
dfr=pd.DataFrame(dictionary)
dfr = dfr.iloc[:-1]
dfr.index= range(1,len(dfr)+1)
styled_df = dfr.style.set_properties(**{'text-align': 'left'}, subset=['Movie Name'])
styled_df
print(dfr)
```
